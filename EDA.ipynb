```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados
url = "https://raw.githubusercontent.com/caioooooo3/Desafio_Lighthouse/main/teste_indicium_precificacao.csv"
data = pd.read_csv(url)

# Viso geral dos dados
data.info()
data.describe()

# Remover outliers
data = data[data['price'] <= 1000]

# Anlise de correlaes
plt.figure(figsize=(12, 6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de correlao entre variveis")
plt.show()

# Anlise de preo mdio por tipo de quarto
data.groupby('room_type')['price'].mean().plot(kind='bar', color='teal')
plt.title("Preo mdio por tipo de quarto")
plt.ylabel("Preo mdio ($)")
plt.show()
```
